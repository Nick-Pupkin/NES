```python
df_GDP = {
    "year" : list(range(1992, 1997)),
    "app_q" : [100, 102, 103, 104, 105],
    "app_p" : [0.25, 0.28, 0.32, 0.34, 0.36],
    "or_q" : [50, 55, 60, 65, 70],
    "or_p" : [0.50, 0.48, 0.45, 0.44, 0.42]
                   }
df_GDP = pd.DataFrame(df_GDP) 
df_GDP
```

```python
# Create an empty DataFrame with empty cells and appropriate size
LI_q_df = pd.DataFrame(columns=["base_year", "year", "LI_q"])
```

```python
k = 0  #row counter
for c_base_year, v_base_year in enumerate(range(1992,1997)):  
    for c_year, v_year in enumerate(range(1993,1997)):
        # LI_q_df = LI_q_df.append({"base_year": v_base_year, "year": v_year}, ignore_index=True) #will be unsupported soon
        LI_q_df = pd.concat([LI_q_df,\
                             pd.DataFrame.from_records([{"base_year": v_base_year, "year": v_year}])])
        
for c_base_year, v_base_year in enumerate(range(1992,1997)):  
    for c_year, v_year in enumerate(range(1993,1997)):
        LI_q_df["LI_q"].values[k] =\
        ((df_GDP.loc[df_GDP["year"] == v_base_year, "app_p"].values[0]\
          * df_GDP.loc[df_GDP["year"] == v_year, "app_q"].values[0]+\
          
          df_GDP.loc[df_GDP["year"] == v_base_year, "or_p"].values[0]*\
          df_GDP.loc[df_GDP["year"] == v_year, "or_q"].values[0])/\
         
         (df_GDP.loc[df_GDP["year"] == v_base_year, "app_p"].values[0]*\
          df_GDP.loc[df_GDP["year"] == (v_year-1), "app_q"].values[0] +\
          
          df_GDP.loc[df_GDP["year"] == v_base_year, "or_p"].values[0]*\
          df_GDP.loc[df_GDP["year"] == (v_year-1), "or_q"].values[0]) - 1)\
        *100
        k += 1
LI_q_df = LI_q_df.reset_index(drop = True)
```
