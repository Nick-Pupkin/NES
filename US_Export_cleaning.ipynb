```python
import warnings
warnings.filterwarnings("ignore")


import pandas as pd
import numpy as np
from tqdm  import tqdm
```

Data from World Bank cleaning ( USD export/gdp %)

```python
data = pd.read_excel(r"C:\Users\Popov\Documents\Research\Hosni\Family_economics\Project\Data_US_trade_GDP.xlsx.")
```

```python
data = data.drop([1, 2, 3, 4, 5], axis=0)
data = data.drop(data.columns[[1, 2, 3]], axis=1)

# Step 1: Extract the years from the column names
data.columns = [col.split()[0] if col != 'Series Name' else col for col in data.columns]

# Step 2: Convert the year columns to integers (except the 'Series Name' column)
new_columns = []
for col in data.columns:
    if col != 'Series Name':
        new_columns.append(int(col))
    else:
        new_columns.append(col)

data.columns = new_columns

# Convert column names to a DataFrame
column_row = pd.DataFrame([data.columns], columns=data.columns)

# Append the original DataFrame after the column row
data = pd.concat([column_row, data], ignore_index=True)

data = data.drop(0, axis=0)
```

```python
data = data.T
data = data.reset_index()
data.columns = ["year",'Exp_World_']
data = data.drop(0, axis=0)
data['Exp_World_'] = pd.to_numeric(data['Exp_World_'], errors='coerce')
data['year'] = data['year'].astype(int)
```
